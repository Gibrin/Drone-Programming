## Решение задачи №2
Для того, чтобы решить задачу, нужно вычислить числа Фибоначчи до заданного количества qty. Числа Фибоначчи — это последовательность, где каждое следующее число является суммой двух предыдущих. Начинается она с 0 и 1.

Вот как это можно сделать:

```python
def get_fib_numbers(qty):
    fib_numbers = [0, 1]  # Начинаем с первых двух чисел
    while len(fib_numbers) < qty:  # Пока количество чисел меньше qty
        fib_numbers.append(fib_numbers[-1] + fib_numbers[-2])  # Добавляем следующее число
    return fib_numbers[:qty]  # Возвращаем только qty чисел

# Пример использования:
fib_numbers = get_fib_numbers(10)
assert len(fib_numbers) == 10
print(fib_numbers)
```

## Объяснение кода:
1. Мы начинаем с двух чисел Фибоначчи: 0 и 1, которые уже находятся в списке fib_numbers.
2. В цикле while продолжаем добавлять новое число, которое равно сумме двух последних чисел в списке, до тех пор, пока не получим требуемое количество чисел.
3. Когда мы достигли нужного количества чисел (проверка через len(fib_numbers) < qty), возвращаем первые qty чисел из списка (через fib_numbers[:qty]).

## Решение задачи 3

Для реализации игры FizzBuzz, нужно создать функцию fizz_buzz(max_number), которая будет проверять каждое число от 1 до max_number на соответствие заданным условиям. Если число делится на 3, мы выводим "fizz". Если число делится на 5, выводим "buzz". Если оно делится и на 3, и на 5, выводим "fizz buzz". Иначе выводим само число.

```python
def fizz_buzz(max_number):
    if max_number < 1:
        print("Максимальное число должно быть не меньше 1")
        return
    
    for i in range(1, max_number + 1):
        if i % 3 == 0 and i % 5 == 0:
            print("fizz buzz")
        elif i % 3 == 0:
            print("fizz")
        elif i % 5 == 0:
            print("buzz")
        else:
            print(i)

def main():
    try:
        max_number = int(input("Введите максимальное число: "))
        fizz_buzz(max_number)
    except ValueError:
        print("Ошибка: введите целое число.")

main()
```

## Объяснение кода:

1. Проверка минимального значения: Если max_number меньше 1, функция выводит сообщение и завершает выполнение, так как игра не может быть проведена с такими данными.

2. Цикл: С помощью цикла for перебираются все числа от 1 до max_number (включительно).

3. Проверка условий:
   - Проверяем, делится ли число одновременно на 3 и на 5, чтобы вывести "fizz buzz".
   - Если число делится только на 3, выводим "fizz".
   - Если число делится только на 5, выводим "buzz".
   - Если ни одно из условий не выполнено, выводим само число.

4. Обработка ошибок: Используется конструкция try...except для проверки корректности ввода данных пользователем, чтобы обрабатывать исключения, возникающие, если введено нецелое число.

Таким образом, программа сначала проверяет ввод пользователя, а затем проходит по каждому числу в заданном диапазоне, выводя результат в соответствии с правилами FizzBuzz.

Исторически решением треугольников называется задача отыскания всех остальных сторон и углов треугольника помимо известных. В этой задаче на вход подаются три стороны треугольника. Найдите все углы треугольника, а также его периметр и площадь. Обратите внимание, что сначала необходимо проверить, образуют ли эти три стороны треугольник.

Обратите внимание, что угол, вычисляемый при помощи арккосинуса, представлен в радианах и его необходимо преобразовать в градусы.

Необходимые математические функции (корень, тригонометрические функции, постоянная пи) можно найти в пакете math.

## Решение задачи 4

Для решения задачи нам потребуются основные математические функции из модуля math. Сначала проверим, могут ли заданные стороны образовать треугольник, затем найдем углы с использованием теоремы косинусов, а также вычислим периметр и площадь по формуле Герона.

```python
import math

def is_triangle(a, b, c):
    return a + b > c and a + c > b and b + c > a

def angle(a, b, c):
    # Используем теорему косинусов для нахождения угла напротив стороны a
    cos_A = (b**2 + c**2 - a**2) / (2 * b * c)
    # Преобразуем значение арккосинуса в градусы
    return math.degrees(math.acos(cos_A))

def triangle_properties(a, b, c):
    if not is_triangle(a, b, c):
        print("Невозможно образовать треугольник с заданными сторонами.")
        return

    # Вычисление углов
    angle1 = angle(a, b, c)
    angle2 = angle(b, a, c)
    angle3 = angle(c, a, b)

    # Периметр
    perimeter = a + b + c

    # Площадь по формуле Герона
    s = perimeter / 2
    area = math.sqrt(s * (s - a) * (s - b) * (s - c))

    # Вывод результатов
    print(f"Угол 1 = {angle1:.2f}")
    print(f"Угол 2 = {angle2:.2f}")
    print(f"Угол 3 = {angle3:.2f}")
    print(f"Периметр треугольника = {perimeter:.3f}")
    print(f"Площадь треугольника = {area:.3f}")

def main():
    try:
        a = float(input("Введите сторону 1: "))
        b = float(input("Введите сторону 2: "))
        c = float(input("Введите сторону 3: "))
        triangle_properties(a, b, c)
    except ValueError:
        print("Ошибка: введите числовое значение для стороны.")

main()
```

## Объяснение кода:
1. Проверка возможности существования треугольника: Используя неравенство треугольника, определяем, может ли существовать треугольник с заданными сторонами. Треугольник существует, если сумма длин любых двух сторон больше длины третьей стороны.

2. Вычисление углов: С помощью теоремы косинусов находим косинус угла напротив каждой стороны и переводим результат из радиан в градусы, так как стандартная функция acos возвращает радианы.

3. Периметр: Сумма всех трех сторон дает периметр.

4. Площадь: Для вычисления площади используем формулу Герона, которая применяется, когда известны стороны треугольника.

5. Обработка ошибок: Программа проверяет, что пользователь ввел числовые значения для сторон и обрабатывает возможные исключения.

Таким образом, программа вычисляет углы треугольника, периметр и площадь, следуя важным геометрическим правилам и формулам.

Для хранения оценок может оказаться полезным использование списка словарей: каждый словарь хранит оценки только одного студента, а ключи в словаре — это названия предметов.

## Решение задачи 5

Для решения задачи мы будем использовать несколько функций для организации ввода данных и расчета статистик. Будем использовать словари для хранения оценок студентов и списки для хранения предметов и студентов.

```python
def read_subjects():
    subjects = []
    while True:
        subject = input("Введите предмет: ").strip().lower()
        if subject == "стоп":
            break
        subjects.append(subject)
    return subjects

def read_students():
    students = []
    while True:
        student = input("Введите фамилию и имя студента через пробел: ").strip().lower()
        if student == "стоп":
            break
        surname, name = student.split(maxsplit=1)
        students.append((surname, name))
    return students

def read_scores(students, subjects):
    scores = []
    for surname, name in students:
        student_scores = {}
        print(f"\nОценка для студента {name.title()} {surname.title()}:")
        for subject in subjects:
            while True:
                try:
                    score = int(input(f"{subject.title()}: ").strip())
                    if score < 1 or score > 10:
                        raise ValueError
                    student_scores[subject] = score
                    break
                except ValueError:
                    print("Ошибка: введите числовое значение от 1 до 10.")
        scores.append(student_scores)
    return scores

def calculate_statistics(students, scores, subjects):
    best_student_index, max_avg_score = 0, 0.0

    # Вычисление средней оценки для каждого студента
    for i, student_scores in enumerate(scores):
        avg_score = sum(student_scores.values()) / len(subjects)
        if avg_score > max_avg_score:
            max_avg_score = avg_score
            best_student_index = i

    best_student = students[best_student_index]

    # Вычисление среднего балла по каждому предмету
    subject_averages = {}
    for subject in subjects:
        total_score = sum(score[subject] for score in scores)
        subject_averages[subject] = total_score / len(students)

    return (best_student, max_avg_score), subject_averages

def main():
    subjects = read_subjects()
    students = read_students()
    scores = read_scores(students, subjects)
    
    (best_student, best_avg_score), subject_averages = calculate_statistics(students, scores, subjects)

    # Вывод результатов
    print("\nСтатистика:")
    print(f"Студент с лучшей успеваемостью\n{best_student[1].title()} {best_student[0].title()}: {best_avg_score:.2f}")
    
    print("\nСредний балл по предмету")
    for subject, avg_score in subject_averages.items():
        print(f"{subject.title()}: {avg_score:.2f}")

main()
```

## Объяснение кода:
1. Функция read_subjects: Используется для ввода предметов, пока не будет введено слово "стоп". Предметы сохраняются в списке.

2. Функция read_students: Вводит студентов в виде строки "Фамилия Имя" и сохраняет их в список кортежей, форматируя имена и фамилии в нижний регистр для унификации.

3. Функция read_scores: Запрашивает оценки по каждому предмету для каждого студента. Оценки хранятся в виде списка словарей, где ключи — это названия предметов, значения — оценки.

4. Функция calculate_statistics: Подсчитывает средние оценки для каждого студента и находит студента с наивысшей средней успеваемостью, а также вычисляет средний балл по каждому предмету.

5. Основная функция main: Оркестрирует вызов функций для ввода данных и печати статистики.

## Решение задачи 6

Метод «Решето Эратосфена» эффективно находит все простые числа до заданного максимального числа. Затем мы выведем их, группируя по десяткам.

```python
def sieve_of_eratosthenes(max_num):
    """Метод 'Решето Эратосфена' для нахождения всех простых чисел до max_num."""
    is_prime = [True] * (max_num + 1)
    p = 2
    while (p * p <= max_num):
        if is_prime[p]:
            for i in range(p * p, max_num + 1, p):
                is_prime[i] = False
        p += 1
        
    primes = [p for p in range(2, max_num + 1) if is_prime[p]]
    
    return primes

def print_primes_by_decades(primes):
    """Вывод простых чисел с разбиением по десяткам."""
    current_decade_start = 0
    decade_primes = []
    
    for prime in primes:
        if prime // 10 > current_decade_start:
            if decade_primes:
                print(", ".join(map(str, decade_primes)) + ",")
            current_decade_start = prime // 10
            decade_primes = []
        decade_primes.append(prime)
    
    if decade_primes:
        print(", ".join(map(str, decade_primes)) + ",")

def main():
    max_num = int(input("Введите максимальное число: "))
    if max_num > 1000:
        max_num = 1000
    
    primes = sieve_of_eratosthenes(max_num)
    print("Простые числа:")
    print_primes_by_decades(primes)

# Запуск программы
main()
```

# Объяснение кода:

1. Функция sieve_of_eratosthenes(max_num):
  - Использует метод «Решето Эратосфена» для нахождения всех простых чисел до max_num.
  - Создает список is_prime, где индексы обозначают числа, а значения указывают, простое ли число.
  - Перебирает числа начиная с 2. Для каждого простого числа помечает все его кратные как не простые.

2. Функция print_primes_by_decades(primes):
  - Разбивает и выводит простые числа по десяткам.
  - Перебирает все простые числа и группирует их, используя целочисленное деление для определения текущего десятка.

3. Функция main():
  - Запрашивает ввод от пользователя.
  - Ограничивает max_num значением 1000, чтобы соблюсти условие задачи.
  - Вызывает функции для нахождения и вывода простых чисел.

## Решение задачи 7

В этой задаче мы будем использовать Python для обработки текста из файла "war_and_peace.txt". Мы соберем статистику по символам и найдём предложения, содержащие формы имени "Андрей".

Первым шагом будет вычисление частоты встречаемости каждого символа в тексте.

```python
import re
from collections import defaultdict

def analyze_text(file_path):
    with open(file_path, 'r', encoding='cp1251') as file:
        text = file.read()
        
    total_chars = len(text)
    char_count = defaultdict(int)
    
    for char in text:
        char_count[char] += 1

    frequency = {char: count / total_chars for char, count in char_count.items()}

    print("Статистика:")
    for char, freq in sorted(frequency.items(), key=lambda x: x[1], reverse=True):
        print(f"{char} : {freq:.2f}")

    return text

def find_sentences_with_andrei(text):
    sentences = re.split(r'\.\s*', text)
    andrei_sentences = [sentence.strip() for sentence in sentences if re.search(r'\bАндрей\b|\bАндрея\b|\bАндрею\b|\bАндреем\b|\bАндрее\b', sentence)]

    print("\nПредложения со словом «Андрей»:")
    for sentence in andrei_sentences[:10]:
        print(sentence + ".")

def main():
    file_path = 'war_and_peace.txt'
    text = analyze_text(file_path)
    find_sentences_with_andrei(text)

main()
```

## Объяснение кода:

1. Функция analyze_text: Читает файл и подсчитывает количество каждого символа с помощью defaultdict. Далее вычисляется частота каждого символа как отношение его количества к общему количеству символов.

2. Функция find_sentences_with_andrei: Разбивает текст на предложения с использованием регулярного выражения, по точкам с возможными пробелами. Затем ищет предложения, содержащие любую форму имени "Андрей", используя регулярные выражения.

3. Главная часть программы main: Инициирует чтение и анализ текста, а также поиск предложений, используя соответствующие функции.

# Задача 8. Калькулятор

Сделайте калькулятор, который поддерживает четыре операции: сложение (+), умножение (×), вычитание (−), деление (÷).

Калькулятор должен поддерживать операции с целыми числами и числами с плавающей точкой, а также должен быть толерантен к пробелам, то есть между операндами и числами может быть неограниченное число пробелов.


Как работает калькулятор
Вы указываете имя файла, содержащего выражения, которые нужно вычислить. Одно выражение — одна строка. Если в выражении есть ошибки, калькулятор запоминает номера строк, в которых они встретились. 

Калькулятор выводит результаты расчётов, номер строки и результат вычисления, а также количество ошибок, номера строк, в которых они встретились, и тип ошибок. 

Входные данные
exprs.txt
2 + 3
-5 * 4
a+ 5

Вывод программы
results.txt
1 5
2 -20

errors.txt
3 name 'a' is not defined

В самом простом варианте можно использовать функцию eval(str) для исполнения выражения, находящегося в строке str. Также можно использовать функцию parse_expr(str) из пакета sympy, с которой мы подробнее познакомимся в следующих модулях.

## Решение задачи 8

В этой задаче мы создадим простой калькулятор, который будет считывать выражения из файла, выполнять расчёты и обрабатывать возможные ошибки. Результаты будут записаны в один файл, а ошибки — в другой.

```python
def read_expressions(file_path):
    with open(file_path, 'r', encoding='utf-8') as file:
        return file.readlines()

def evaluate_expression(expression, line_number, results, errors):
    try:
        # Удаляем пробелы из начала и конца и вычисляем выражение
        result = eval(expression.strip())
        results.append(f"{line_number} {result}")
    except Exception as e:
        errors.append(f"{line_number} {str(e)}")

def write_to_file(file_path, lines):
    with open(file_path, 'w', encoding='utf-8') as file:
        for line in lines:
            file.write(line + '\n')

def main():
    input_file = 'exprs.txt'
    results_file = 'results.txt'
    errors_file = 'errors.txt'

    expressions = read_expressions(input_file)
    results = []
    errors = []

    for line_number, expression in enumerate(expressions, start=1):
        evaluate_expression(expression, line_number, results, errors)

    write_to_file(results_file, results)
    write_to_file(errors_file, errors)

    print("\nРезультаты записаны в файл results.txt.")
    print(f"Число ошибок: {len(errors)}. Смотрите файл errors.txt для деталей.")

main()
```

### Пояснение

1. Функция read_expressions: Считывает все строки из входного файла с выражениями.

2. Функция evaluate_expression: 
   - Для каждого выражения вызывает eval, чтобы рассчитать результат.
   - Обрабатывает любые возникающие при этом ошибки (например, отсутствующие операторы или недоступные переменные) и сохраняет сообщения об ошибках с номерами строк.

3. Функция write_to_file: Записывает результаты и ошибки в соответствующие файлы.

4. Главная часть программы main:
   - Читает выражения, вычисляет результаты и определяет ошибки.
   - Результаты и ошибки записываются в отдельные файлы, и производится вывод информации о выполнении программы на экран.
