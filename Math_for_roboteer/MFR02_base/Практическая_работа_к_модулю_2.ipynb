## Задача 1
Решить систему линейных алгебраических уравнений как систему уравнений и в матричном виде:

$
    \left\{\begin{matrix}
       3x + 4y -2z - 6d = 3 \\
       3y -2z +x  + 7d= -2 \\
       5x - 7z -2y -4d = 5 \\
       2x + y - 3z - 4d = 6
       \end{matrix}\right.
$

```python
# 1. Решение системы уравнений
from sympy import symbols, Eq, solve
x,y,z,d=symbols('x y z d')
eqs=[
    Eq(3*x+4*y-2*z-6*d,3),
    Eq(x+3*y-2*z+7*d,-2),
    Eq(5*x-2*y-7*z-4*d,5),
    Eq(2*x+y-3*z-4*d,6)
]
solution=solve(eqs,(x,y,z,d))
print("Решение системы уравнений:",solution)

# 2. Решение матричным методом
import numpy as np
A=np.array([[3,4,-2,-6],
            [1,3,-2,7],
            [5,-2,-7,-4],
            [2,1,-3,-4]])
B=np.array([3,-2,5,6])
matrix_solution=np.linalg.solve(A,B)
print("Решение матричным методом:",dict(zip(['x','y','z','d'],matrix_solution)))
```

## Задача 2
Построить матрицу с необходимыми свойствами или объяснить, почему это невозможно:

1. Пространство колонок состоит из $\begin{bmatrix}1\\1\\0\end{bmatrix},\ \begin{bmatrix}0\\0\\1\end{bmatrix}$, пространство строчек содержит $\begin{bmatrix}1\\2\end{bmatrix},\ \begin{bmatrix}2\\5\end{bmatrix}$.

2. Пространство колонок состоит из $\begin{bmatrix}1\\1\\3\end{bmatrix}$, ядро (nullspace) — $\begin{bmatrix}3\\1\\1\end{bmatrix}$.

3. Ядро состоит из $\begin{bmatrix}1\\3\end{bmatrix}$, пространство колонок — из $\begin{bmatrix}3\\1\end{bmatrix}$.

Подсказка: «содержит» означает, что результат может быть получен из других векторов с помощью линейной комбинации.

```python
from sympy import Matrix

def in_span(vec, basis):
    """Проверяет, лежит ли vec в линейной оболочке списка векторов basis."""
    if not basis:
        return all(v == 0 for v in vec)
    # Собираем матрицу из базисных векторов как столбцов
    M = Matrix.hstack(*basis) if len(basis) > 1 else Matrix(basis[0]).reshape(len(basis[0]), 1)
    # Решаем M * x = vec
    try:
        # Используем solve_least_squares или проверку совместности
        # Более надёжно: ранг [M | vec] == ранг M
        aug = M.row_join(vec.reshape(M.rows, 1))
        return M.rank() == aug.rank()
    except Exception:
        return False

print("="*60)
print("ЗАДАЧА 1")
print("="*60)

A1 = Matrix([[1, 0],
             [1, 0],
             [0, 1]])

# Проверка строкового пространства
row_basis = A1.rowspace()  # список строк-векторов (1×2)
v1 = Matrix([1, 2])
v2 = Matrix([2, 5])

# Преобразуем строки из rowspace в столбцы для удобства, но лучше работать с транспонированием
# Однако in_span ожидает векторы той же формы — сделаем их столбцами
row_basis_col = [r.T for r in row_basis]  # теперь каждый — столбец 2×1

print("Col(A) = span{[1,1,0], [0,0,1]}?", 
      set(map(tuple, A1.columnspace())) == {tuple(Matrix([1,1,0])), tuple(Matrix([0,0,1]))})

print("[1,2] ∈ Row(A)?", in_span(v1, row_basis_col))
print("[2,5] ∈ Row(A)?", in_span(v2, row_basis_col))

print("\n" + "="*60)
print("ЗАДАЧА 2")
print("="*60)
print("Невозможно: rank=1, nullity=1 ⇒ n=2, но ядро задано в R^3 ⇒ n=3 → противоречие.")

print("\n" + "="*60)
print("ЗАДАЧА 3")
print("="*60)

B = Matrix([[-9, 3],
            [-3, 1]])

col_target = Matrix([3, 1])
null_target = Matrix([1, 3])

# Проверка: каждый столбец B лежит в span{col_target}
cols_ok = all(in_span(B.col(i), [col_target]) for i in range(B.cols))

# Проверка: null_target в ядре
null_ok = B * null_target == Matrix([0, 0])

print("Столбцы в span{[3,1]}?", cols_ok)
print("B @ [1,3] = 0?", null_ok)
print("Ранг B =", B.rank())
```

```python
from sympy import Matrix, symbols, eye

# Матрица A (3x4)
A = Matrix([[1, 2, 0, 0],
            [2, 5, 0, 0],
            [0, 0, 2, 0]])

print("Матрица A:")
print(A)
print()

# 1) Ранг
rank = A.rank()
print(f"1) Ранг матрицы: {rank}")

# 2) Размерность ядра
n = A.shape[1]  # число столбцов
nullity = n - rank
print(f"2) Размерность ядра: {nullity}")

# 3) Базис пространства колонок
col_basis = A.columnspace()
print("3) Базис пространства колонок:")
for i, vec in enumerate(col_basis):
    print(f"   v{i+1} = {vec}")

# 4) Базис ядра
null_basis = A.nullspace()
print("4) Базис ядра:")
if null_basis:
    for i, vec in enumerate(null_basis):
        print(f"   n{i+1} = {vec}")
else:
    print("   Тривиальное ядро {0}")

# 5) RREF
rref_A, pivots = A.rref()
print("5) RREF матрицы:")
print(rref_A)

# 6) Решение уравнения A*x = b
b1, b2, b3 = symbols('b1 b2 b3')
b = Matrix([b1, b2, b3])

# Расширенная матрица
aug = A.row_join(b)
rref_aug, _ = aug.rref()
print("\n6) RREF расширенной матрицы [A|b]:")
print(rref_aug)

# Решение системы
print("\n7) Решение системы A*x = b:")
print("Из RREF видим:")
print("x1 = свободная переменная?")
print("x2 = свободная переменная?")
print("x3 = b3/2")
print("x4 = свободная переменная (t)")

# Найдем частное решение при свободных переменных = 0
# Из RREF: x1 = ?, x2 = ?, x3 = b3/2, x4 = 0
# Но нужно решить правильно

# Решаем символьно:
x1, x2, x3, x4 = symbols('x1 x2 x3 x4')
# Уравнения:
eq1 = x1 + 2*x2 - b1
eq2 = 2*x1 + 5*x2 - b2
eq3 = 2*x3 - b3

from sympy import solve
solution = solve([eq1, eq2, eq3], [x1, x2, x3])
print("\n8) Символьное решение (x4 - свободная):")
print(f"x1 = {solution[x1]}")
print(f"x2 = {solution[x2]}")
print(f"x3 = {solution[x3]}")
print(f"x4 = t (произвольная константа)")

# Частное решение при t = 0
print("\n9) Частное решение (при x4 = 0):")
x_part = Matrix([solution[x1], solution[x2], solution[x3], 0])
print(f"x_part = {x_part}")

# Общее решение = частное + линейная комбинация векторов ядра
print("\n10) Общее решение:")
print(f"x = x_part + t * n1, где n1 - базис ядра")
if null_basis:
    print(f"   = {x_part} + t * {null_basis[0]}")

# Проверка
print("\n11) Проверка A*x_part = b:")
print(f"A * x_part = {A * x_part}")
print(f"b = {b}")
print(f"Равны? {A * x_part == b}")
```

## Задача 1
Спроецируйте вектор $b=\begin{bmatrix}3\\4\\4\end{bmatrix}$ вначале на прямую $a=\begin{bmatrix}2\\2\\1\end{bmatrix}$, а затем на плоскость, содержащую ещё вектор $a^* = \begin{bmatrix}1\\0\\0\end{bmatrix}$. Проверьте, что вектор ошибки перпендикулярен $a$ (*подсказка: скалярное произведение*) и что второй вектор ошибки также перпендикулярен $a^*$.

*Для проверки себя используйте GeoGebra или Desmos*.

```python
import numpy as np

# Исходные векторы
b = np.array([3, 4, 4], dtype=float)
a = np.array([2, 2, 1], dtype=float)
a_star = np.array([1, 0, 0], dtype=float)

print("1. Проекция b на прямую вдоль a:")
# Проекция b на a: (a·b)/(a·a) * a
proj_a = (np.dot(a, b) / np.dot(a, a)) * a
print(f"Проекция на a: {proj_a}")

# Вектор ошибки для первой проекции
e1 = b - proj_a
print(f"\nВектор ошибки e1 = b - проекция_на_a: {e1}")
print(f"Проверка перпендикулярности (a·e1): {np.dot(a, e1)}")
print(f"Должно быть 0 (с точностью до погрешности): {np.abs(np.dot(a, e1)) < 1e-10}")

print("\n" + "="*50)
print("2. Проекция b на плоскость, натянутую на a и a*:")
# Матрица A, чьи столбцы - векторы a и a*
A = np.column_stack([a, a_star])  # 3×2 матрица

# Проекция на плоскость: A * (A^T * A)^(-1) * A^T * b
AT = A.T
ATA = AT @ A
# Проверяем, что ATA обратима (столбцы линейно независимы)
if np.linalg.det(ATA) != 0:
    ATA_inv = np.linalg.inv(ATA)
    proj_plane = A @ ATA_inv @ AT @ b
    print(f"Проекция на плоскость: {proj_plane}")
    
    # Вектор ошибки для второй проекции
    e2 = b - proj_plane
    print(f"\nВектор ошибки e2 = b - проекция_на_плоскость: {e2}")
    print(f"Проверка перпендикулярности e2 к a: {np.dot(a, e2)}")
    print(f"Проверка перпендикулярности e2 к a*: {np.dot(a_star, e2)}")
    print(f"Должны быть 0: {np.abs(np.dot(a, e2)) < 1e-10} и {np.abs(np.dot(a_star, e2)) < 1e-10}")
else:
    print("Столбцы линейно зависимы!")
```

2. Найдите суммарную ошибку и сравните.

```python
import numpy as np
import matplotlib.pyplot as plt

# Данные
x = np.array([-1, 1, -1, 2, 0, 1])
y = np.array([2, 0, 0, 1, -1, 2])

# 1. Решение для прямой: y = a0 + a1*x
A_line = np.column_stack([np.ones_like(x), x])
coeff_line = np.linalg.inv(A_line.T @ A_line) @ A_line.T @ y
a0, a1 = coeff_line
y_line = a0 + a1 * x
SSE_line = np.sum((y - y_line)**2)

# 2. Решение для параболы: y = b0 + b1*x + b2*x²  
A_parab = np.column_stack([np.ones_like(x), x, x**2])
coeff_parab = np.linalg.inv(A_parab.T @ A_parab) @ A_parab.T @ y
b0, b1, b2 = coeff_parab
y_parab = b0 + b1 * x + b2 * x**2
SSE_parab = np.sum((y - y_parab)**2)

# Визуализация
x_smooth = np.linspace(-1.5, 2.5, 100)
plt.scatter(x, y, c='red', s=50, label='Данные')
plt.plot(x_smooth, a0 + a1*x_smooth, 'b-', label=f'Прямая: y={a0:.3f}+{a1:.3f}x')
plt.plot(x_smooth, b0 + b1*x_smooth + b2*x_smooth**2, 'g-', 
         label=f'Парабола: y={b0:.3f}{b1:+.3f}x{b2:+.3f}x²')
plt.grid(True, alpha=0.3); plt.legend(); plt.show()

# Вывод результатов
print(f"Прямая: y = {a0:.4f} + {a1:.4f}x, SSE = {SSE_line:.4f}")
print(f"Парабола: y = {b0:.4f} + {b1:.4f}x + {b2:.4f}x², SSE = {SSE_parab:.4f}")
print(f"Улучшение: {SSE_line-SSE_parab:.4f} ({100*(SSE_line-SSE_parab)/SSE_line:.1f}%)")
```

## Задача 1
Имеется система координат $A$ и координаты вектора $^AP$ относительно $A$. Потом мы трансформировали $^AP$, вначале повернув его вокруг $Z_A$ на угол $\theta$, а потом вокруг $Y_A$ на угол $\gamma$. Определите матрицу поворота $R(\theta, \gamma)$.


```python
import sympy as sp

# Определение символьных переменных
θ, γ = sp.symbols('θ γ')

# 1. Создаем матрицу поворота вокруг оси Z на угол θ
Rz = sp.Matrix([
    [sp.cos(θ), -sp.sin(θ), 0],
    [sp.sin(θ),  sp.cos(θ), 0],
    [0,          0,         1]
])

# 2. Создаем матрицу поворота вокруг оси Y на угол γ
Ry = sp.Matrix([
    [sp.cos(γ),  0, sp.sin(γ)],
    [0,          1, 0],
    [-sp.sin(γ), 0, sp.cos(γ)]
])

# 3. Вычисляем произведение: R(θ, γ) = Ry(γ) × Rz(θ)
R = Ry @ Rz

# 4. Упрощаем результат
R_simplified = sp.simplify(R)

print("Матрица поворота R(θ, γ) = Ry(γ) × Rz(θ) (вычисленная):")
sp.pprint(R_simplified)
```

```python
import numpy as np
import sympy as sp

# Упрощенная модель: 3 вращательных звена
θ1, θ2, θ3 = sp.symbols('θ1 θ2 θ3')
L1, L2, L3 = sp.symbols('L1 L2 L3')  # Длины трех звеньев

# Единая гомогенная матрица для каждого звена
def T_link(theta, L):
    """Матрица преобразования для одного звена: поворот + сдвиг"""
    return sp.Matrix([
        [sp.cos(theta), -sp.sin(theta), L * sp.cos(theta)],
        [sp.sin(theta),  sp.cos(theta), L * sp.sin(theta)],
        [0,             0,              1]
    ])

# Последовательность преобразований
T01 = T_link(θ1, L1)  # Звено 1
T12 = T_link(θ2, L2)  # Звено 2
T23 = T_link(θ3, L3)  # Звено 3

# Полная матрица из СК0 в СК3 (конечную)
T03 = sp.simplify(T01 * T12 * T23)

print("Матрица преобразования T₀³:")
sp.pprint(T03)

# В виде последовательности умножений
print("\nПоследовательность: T₀³ = T_link(θ₁, L₁) × T_link(θ₂, L₂) × T_link(θ₃, L₃)")

# Явные формулы
print("\nПозиция конечной точки (x, y) в системе x₀y₀:")
x = L1*sp.cos(θ1) + L2*sp.cos(θ1+θ2) + L3*sp.cos(θ1+θ2+θ3)
y = L1*sp.sin(θ1) + L2*sp.sin(θ1+θ2) + L3*sp.sin(θ1+θ2+θ3)
print(f"x = {x}")
print(f"y = {y}")
print(f"Угол ориентации: θ = {sp.simplify(θ1 + θ2 + θ3)}")

# Проверка на граничных случаях
print("\n" + "="*60)
print("Проверка (численно, L₁=2, L₂=1.5, L₃=1):")

def numeric_T03(theta1, theta2, theta3, L1=2.0, L2=1.5, L3=1.0):
    """Численное вычисление T03"""
    T01_n = np.array([[np.cos(theta1), -np.sin(theta1), L1*np.cos(theta1)],
                      [np.sin(theta1),  np.cos(theta1), L1*np.sin(theta1)],
                      [0, 0, 1]])
    T12_n = np.array([[np.cos(theta2), -np.sin(theta2), L2*np.cos(theta2)],
                      [np.sin(theta2),  np.cos(theta2), L2*np.sin(theta2)],
                      [0, 0, 1]])
    T23_n = np.array([[np.cos(theta3), -np.sin(theta3), L3*np.cos(theta3)],
                      [np.sin(theta3),  np.cos(theta3), L3*np.sin(theta3)],
                      [0, 0, 1]])
    return T01_n @ T12_n @ T23_n

# Простые проверки
cases = [
    ("Все углы 0°", 0, 0, 0),
    ("Только θ₁=90°", np.pi/2, 0, 0),
    ("Только θ₂=90°", 0, np.pi/2, 0),
    ("Все углы 90°", np.pi/2, np.pi/2, np.pi/2)
]

for name, t1, t2, t3 in cases:
    T = numeric_T03(t1, t2, t3)
    x_val, y_val = T[0, 2], T[1, 2]
    angle = np.arctan2(T[1, 0], T[0, 0])
    print(f"\n{name}:")
    print(f"  Позиция: ({x_val:.2f}, {y_val:.2f})")
    print(f"  Ориентация: {np.degrees(angle):.1f}°")
```
