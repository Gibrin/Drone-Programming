Сохраним эти же данные в другой файл под названием "copy of world-data-2023.csv". Для этого заполним пропуск в следующей ячейке.

Новый файл должен появиться на левой панели рядом с ранее загруженным файлом. Можно скачать его и убедиться, что данные в нём остались те же самые.

```python
import pandas as pd
import os

# Скачиваем файл напрямую
url = "https://drive.google.com/uc?export=download&id=131-TzL7VSM-Qs9oMWhqfuSSyVaMDq3me"
file_name = "world-data-2023.csv"

# Скачиваем файл если его нет
if not os.path.exists(file_name):
    print("Скачиваем файл...")
    import requests
    response = requests.get(url)
    with open(file_name, 'wb') as f:
        f.write(response.content)
    print("Файл успешно скачан!")
else:
    print("Файл уже существует")

# Теперь импортируем данные
data = pd.read_csv(file_name)

# Дальше ваш код...
print("Тип переменной data:", type(data))
data.to_csv("copy of world-data-2023.csv")

print("Текущая рабочая директория:", os.getcwd())
print("Файл создан:", os.path.exists("copy of world-data-2023.csv"))
```

Задание 4. Фильтры и условия

Во многом навигация по структуре DataFrame по названиям столбцов или по индексам реализована во многих других объектах. Однако есть дополнительные функции, позволяющие делать выборки данных по более сложным критериям.

Для примера давайте рассмотрим следующий пример. Запустите ячейку ниже.

data['Life expectancy'] > 83

С помощью оператора сравнения мы выяснили, в каких строках фрейма параметр «Продолжительность жизни» строго больше 83. Если условие выполняется, то само значение заменяется на True. В ином случае — False.

Пойдём дальше и подставим это логическое выражение вместо ключа.

data[data['Life expectancy'] > 83]

Итого мы получили только 5 стран, которые удовлетворяют этому критерию поиска. Попробуем получить более сложное условие.

Найдите страны, в которых средняя продолжительность жизни больше 83, но меньше 84.

Условия фильтрации в DataFrame составляются по тому же принципу, что и обычные булевы условия, за тем исключением, что вместо традиционного оператора «and» используется оператор «&», а вместо «or» — «|». Более полный список булевых операторов для условий можно найти в документации.

#вставьте ваш код здесь

Если вы получили фрейм из трёх строк — Сингапур, Испания и Швейцария, то вы сделали всё абсолютно верно!

```python
data['Life expectancy'] > 83
```

```python
data[data['Life expectancy'] > 83]
```

```python
data[(data['Life expectancy'] > 83) & (data['Life expectancy'] < 84)]
```

Сгруппируем все данные по уникальным значениям столбца Official language, а затем соберём сводные значения по количеству стран, для которых этот язык является официальным.
