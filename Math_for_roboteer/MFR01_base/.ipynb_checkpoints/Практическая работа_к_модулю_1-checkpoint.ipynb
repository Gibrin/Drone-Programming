**Задача 2**

Даны точки $\mathbf{A}(3,-2)$ и $\mathbf{B}(1,4)$. Точка $\mathbf{M}$ лежит на прямой $\mathbf{AB}$ так, что
$|\mathbf{AM}| = 3|\mathbf{AB}|$. Найдите координаты точки $ \mathbf{M}$, если точки $ \mathbf{M}$ и $ \mathbf{B}$ лежат по одну сторону от A.

```python
# Дано
A = (3, -2)
B = (1, 4)

# Вектор AB
AB = (B[0] - A[0], B[1] - A[1])

# Т AM = 3 * AB
AM = (3 * AB[0], 3 * AB[1])

# Координаты точки M
M = (A[0] + AM[0], A[1] + AM[1])
print(M)
```

**Задача 3**

Все три вектора $\vec{a}, \vec{b},\vec{c}$ имеют длину $3$, и $\vec{a} + \vec{b} + \vec{c} = \vec{0}$.

Найдите результат вычисления: $\vec{a} \cdot \vec{b} + \vec{b} \cdot \vec{c} + \vec{c} \cdot \vec{a}$.

*Подсказка: необходимо нарисовать этот случай (треугольник)*.

```python
length = 3

angle = sp.rad(120)  # или 2*sp.pi/3

dot_product = length**2 * sp.cos(angle)

result = 3 * dot_product

print(f"Сумма скалярных произведений = {result}")  # -27/2
print(f"Численное значение = {result.evalf()}")     # -13.5
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Углы для трёх векторов
angles = [0, 120, 240]  # градусы
angles_rad = np.deg2rad(angles)

# Векторы длиной 3
vectors = 3 * np.column_stack([np.cos(angles_rad), np.sin(angles_rad)])

fig, ax = plt.subplots()
for i, vec in enumerate(vectors):
    ax.plot([0, vec[0]], [0, vec[1]], 'o-', label=f'Вектор {i+1}')

ax.set_aspect('equal')
ax.grid(True)
ax.legend()
plt.show()
```

**Видео 6**

**Задача 1**

Даны $A=\begin{bmatrix} 2 & -1 & -1\end{bmatrix}$ и $B=\begin{bmatrix} -2 \\ -1 \\ 3 \end{bmatrix}$. \\
    1. Найдите $AB$ и $BA$, если они существуют. \\
    2. Найдите $A^TB$ и $BA^T$, если они существуют.

```python
import numpy as np

A = np.array([[2, -1, -1]])      # 1x3
B = np.array([[-2], [-1], [3]])  # 3x1

print("AB =", (A @ B).item())     # скаляр → -6
print("BA =\n", B @ A)           # 3x3 матрица

# A^T B и B A^T — не существуют (размерности 3x1 и 3x1)
try:
    A.T @ B
except ValueError as e:
    print("A^T B не существует:", e)

try:
    B @ A.T
except ValueError as e:
    print("B A^T не существует:", e)
```

**Задача 2**

Дана матрица $A = \begin{bmatrix} 1 & 4 \\ 9 & 1\end{bmatrix}$.

Найдите $l$, если определитель $A - El$ равен $0$, где $E$ — единичная матрица.

```python
import numpy as np

A = np.array([[1, 4], [9, 1]])

eigenvalues = np.linalg.eigvals(A)
print("Собственные значения (l):", eigenvalues) 
```

**Видео 7**

**Задача 1**

Вычислите ранг матрицы $\begin{bmatrix}1&2\\2&4\\1&1\end{bmatrix}$ с помощью функции $rank()$ и c помощью $rref()$.

```python
import numpy as np

M = np.array([[1, 2],
              [2, 4],
              [1, 1]])

rank = np.linalg.matrix_rank(M)
print("Ранг матрицы (через rank):", rank)
```

```python
import sympy as sp

M = sp.Matrix([[1, 2],
               [2, 4],
               [1, 1]])

rref_matrix, pivot_cols = M.rref()
rank = len(pivot_cols)  # количество ведущих столбцов = ранг

print("Ступенчатый вид (rref):")
sp.pprint(rref_matrix)
print("Ранг матрицы (через rref):", rank)
```

**Видео 8**

**Задача 1**

Решите уравнения:

1) $\begin{bmatrix}2&5\\1&3\end{bmatrix}X=\begin{bmatrix}2&1\\1&1\end{bmatrix}$; \\
2)  $X\begin{bmatrix}2&5\\1&3\end{bmatrix}=\begin{bmatrix}2&1\\1&1\end{bmatrix}$;

```python
import numpy as np

A = np.array([[2, 5],
              [1, 3]])
B = np.array([[2, 1],
              [1, 1]])

# 1.1: AX = B → X = A⁻¹ B
X1 = np.linalg.inv(A) @ B
print("X1 (AX = B):")
print(X1)

# 1.2: XA = B → X = B A⁻¹
X2 = B @ np.linalg.inv(A)
print("\nX2 (XA = B):")
print(X2)
```

**Задача 2**

Найдите значения $AA^+$, $AA^+A$ и $A^+A$, если $A=\begin{bmatrix}3 \\ 7 \\ 2\end{bmatrix}$.

```python
import numpy as np

A = np.array([[3], [7], [2]])

# Псевдообратная
A_pinv = np.linalg.pinv(A)

# AA⁺
AA_pinv = A @ A_pinv
print("AA⁺ =")
print(AA_pinv)

# AA⁺A
AA_pinv_A = AA_pinv @ A
print("\nAA⁺A =")
print(AA_pinv_A)

# A⁺A
A_pinv_A = A_pinv @ A
print("\nA⁺A =")
print(A_pinv_A)
```

**Видео 9**

**Задача 1**

Решите уравнения:

1) $x^2 + 2x + 4 =0$ \\
2) $\bar{x} = 2 + 2x$

```python
# 1.1: x² + 2x + 4 = 0
x = sp.symbols('x')
eq1 = x**2 + 2*x + 4
solutions1 = sp.solve(eq1, x)
print("Корни уравнения 1:", solutions1)

# 1.2: conjugate(x) = 2 + 2x
x = sp.symbols('x', complex=True)
eq2 = sp.conjugate(x) - (2 + 2*x)
solution2 = sp.solve(eq2, x)
print("Решение уравнения 2:", solution2)
```
