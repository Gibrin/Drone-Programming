## Задание 1. Построение линейного графика в Matplotlib

Многие физические процессы называются гармоническими. Они характеризуются колеблющимися величинами, которые изменяются во времени по закону синуса или косинуса и описываются волновыми уравнениями. Примеры подобных зависимостей встречаются в исследованиях механики движения струны, при обработке сигналов и описании процессов в квантовой механике. Давайте разберёмся, как визуализировать такие модели.

Представим движение струны в идеальных условиях: если её вывести из положения равновесия, она начнёт колебаться. Наша задача — визуализировать этот процесс.

Для начала нужно подготовить набор данных, который мы будем визуализировать. Воспользуемся методами библиотеки NumPy, чтобы создать набор из 100 точек переменной $t$ в промежутке $[0, 10\pi]$.

```python
import numpy as np
from math import pi



t_data = np.linspace(0, 10 * np.pi, 100)

print(t_data)
```

Таким образом, мы добавили сетку времени, в которой будем исследовать наш процесс. Теперь определим новую переменную $y$, которая будет соответствовать величине отклонения точки струны от точки равновесия, в которой она находилась перед воздействием.

\begin{equation*}
y = A\cdot \cos t,
\end{equation*}
где $A$ — амплитуда движения струны. Пусть эта величина в нашем примере будет равна трём.

```python
A = 3

y_data = A * np.cos(t_data)
```

Далее построим график этой зависимости с помощью библиотеки Matplotlib. Для этого импортируем её.

```python
import matplotlib.pyplot as plt

plt.plot(t_data, y_data)
```

Очевидно, графику не хватает описания. Давайте подготовим его.

Добавим на график следующие элементы:


*   заголовок графика — «Колебание струны»;
*   подпись горизонтальной оси — «Время, с»;
*   подпись вертикальной оси — «Отклонение, мм»;
*   легенду;
*   сетку.

А также установим границы отображаемой области по вертикальной оси в границах $[−5,5]$.




```python
plt.plot(t_data, y_data, label='y(t)=A cos(t)')

plt.title("Колебание струны")
plt.xlabel("Время, с")
plt.ylabel("Отклонение, мм")
plt.legend()
plt.grid()
plt.ylim(-5, 5)
plt.show()
```

## Задание 4. Построение нескольких графиков

Построим в одном графическом объекте несколько графиков. Первый график будет показывать зависимость отклонения струны от времени, а второй — зависимость энергии колебаний от амплитуды.

В этом случае рассмотрим более реалистичную ситуацию. Так как амплитуда со временем затухает, зависимость будет выглядеть следующим образом:
\begin{equation*}
y(t) = A\cdot e^{-b\cdot t} \cos t
\end{equation*}

Построим набор точек и их график при $b=0,15$. Добавьте на график те же элементы, что и на изображениях выше.

```python
t = np.linspace(0,40,100)

y_real = A * np.exp(-0.15*t)*np.cos(t)



plt.plot(t_data, y_real, label='y (real)')

plt.title("Колебание струны")
plt.xlabel("Время, с")
plt.ylabel("Отклонение, мм")
plt.legend()
plt.grid(True)
plt.ylim(-5,5)
```

Амплитуда затухающих колебаний определяется первыми двумя множителями.
\begin{equation*}
A_t = A\cdot e^{-b\cdot t}
\end{equation*}

Энергию колебаний в упрощённой модели можно соотнести с амплитудой по формуле:
\begin{equation*}
W \sim A_t^2
\end{equation*}

Построим отдельно график зависимости энергии от амплитуды, добавим следующие элементы:
*   заголовок графика — «Энергия колебаний струны»;
*   подпись горизонтальной оси — «Амплитуда, мм»;
*   подпись вертикальной оси — «Энергия»;
*   легенду;
*   сетку.

Обратите внимание, как задаются ограничения для горизонтальной оси. Если сначала указывается наибольшее значение, а затем наименьшее, то ось переворачивается.

```python
ampl_data = A * np.exp(-0.15*t)
energy = ampl_data**2

plt.plot(ampl_data, energy, label='Энергия')
plt.title("Энергия колебаний струны")
plt.xlabel("Амплитуда, мм")
plt.ylabel("Энергия")
plt.legend()
plt.grid()
plt.xlim(3, 0)
plt.ylim(0, 10)
```

Два связанных графика удобнее располагать в одном графическом объекте. Воспользуемся функцией plt.subplots, чтобы создать графические сцены.

Разместите в plot1 график затухающих колебаний, а в plot2 — график энергии. Добавьте к ним подписи y (real) и energy, которые затем появятся в легенде.

```python
fig, (plot1, plot2) = plt.subplots(1 , 2)
plt.subplot(1, 2,1)
plt.plot(t_data, y_real, label='y (real)')
plt.grid()
plt.legend()
plt.xlabel('Time, sec')
plt.ylabel('Real')
plt.subplot(1, 2, 2)
plt.plot(ampl_data, energy, label='Энергия', marker = 11)
plt.grid()
plt.legend()

```

Настроим описание всех элементов в обоих графиках.
Методы настройки описательных элементов отличаются от использованных выше, поэтому приведём их для одного подграфика. Добавьте настройки для графика, показывающего зависимость энергии от амплитуды, по аналогии с первым на основе элементов из одиночного графика.

```python
fig, (plot1, plot2) = plt.subplots(1, 2, figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(t_data, y_real, label='y (real)')
plt.subplot(1, 2, 2)
plt.plot(ampl_data, energy, label='Энергия')

fig.suptitle("Модель колебаний струны")

plot1.set_title("Затухающие колебания")
plot1.set_xlabel("Время, с")
plot1.set_ylabel("Отклонение, мм")
plot1.legend()
plot1.grid(True)
plot1.set_ylim(-5, 5)

#fig.suptitle("Модель колебаний струны")

plot2.set_title("Энергия колебаний струны")
plot2.set_xlabel("Амплитуда, мм")
plot2.set_ylabel("Энергия")
plot2.legend()
plot2.grid(True)
plot2.set_xlim(3, 0)
plot2.set_ylim(0, 10)
# Вставьте свой код здесь (настройка стилей plot2)
```

В результате вы должны получить графический объект следующего вида:

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры
A = 3
b = 0.15
t = np.linspace(0, 40, 100)  # единый массив времени

# Затухающие колебания
y_real = A * np.exp(-b * t) * np.cos(t)

# Амплитуда и энергия
ampl_data = A * np.exp(-b * t)
energy = ampl_data ** 2

# Создаём фигуру с двумя подграфиками
fig, (plot1, plot2) = plt.subplots(1, 2, figsize=(14, 5))

# --- Левый график: затухающие колебания ---
plot1.plot(t, y_real, label='Колебания')
plot1.set_title("Колебание струны")
plot1.set_xlabel("Время, с")
plot1.set_ylabel("Отклонение, мм")
plot1.legend()
plot1.grid(True)
plot1.set_ylim(-5, 5)

# --- Правый график: энергия от амплитуды ---
plot2.plot(ampl_data, energy, label='Энергия', color='orange')
plot2.set_title("Энергия колебаний струны")
plot2.set_xlabel("Амплитуда, мм")
plot2.set_ylabel("Энергия")
plot2.legend()
plot2.grid(True)
plot2.set_xlim(3, 0)   # ← ось X перевёрнута: от 3 к 0
plot2.set_ylim(0, 10)

# Общий заголовок (опционально, но полезно)
fig.suptitle("Модель колебаний струны", fontsize=14)

plt.tight_layout()  # чтобы подписи не налезали
plt.show()
```

В качестве упражнения построим график из задания 3. Для этого нужно создать словарь из данных и воспользоваться двумя функциями из библиотеки Bokeh:


*   line(x, y, source) — построение линейного графика;
*   scatter(x, y, source) — построение графика рассеяния.



```python
from bokeh.io import output_notebook
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
import numpy as np

# Включаем поддержку ноутбука
output_notebook()

# Генерация данных
t_data = np.linspace(0, 10 * np.pi, 100)
A = 3
y_data = A * np.cos(t_data)
noise = np.random.normal(loc=0, scale=0.2, size=100)
y_noise = y_data + noise

# Источник данных
source = ColumnDataSource({
    't': t_data,
    'y': y_data,
    'y_noise': y_noise
})

# График
p = figure(
    title="Колебание струны",
    x_axis_label="Время, с",
    y_axis_label="Отклонение, мм",
    width=800,
    height=400
)

p.scatter(x='t', y='y_noise', source=source, color='green', size=5, legend_label='y (noise)')
p.line(x='t', y='y', source=source, color='blue', legend_label='y (ideal)')
p.legend.location = "top_right"

show(p)
```

Построенный график получился интерактивным за счёт инструментов, расположенных справа от него.

```python
# Установка (обычно уже есть в Colab)
!pip install plotly

import plotly.graph_objects as go
import numpy as np

# Данные
t = np.linspace(0, 10 * np.pi, 100)
A = 3
y_ideal = A * np.cos(t)
y_noise = y_ideal + np.random.normal(0, 0.2, size=t.shape)

# График
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=t, y=y_noise,
    mode='markers',
    marker=dict(color='green', size=5),
    name='y (noise)'
))

fig.add_trace(go.Scatter(
    x=t, y=y_ideal,
    mode='lines',
    line=dict(color='blue'),
    name='y (ideal)'
))

fig.update_layout(
    title="Колебание струны",
    xaxis_title="Время, с",
    yaxis_title="Отклонение, мм",
    width=800,
    height=400
)

fig.show()
```
