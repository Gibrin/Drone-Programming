## Задача 1
Найдите производные функции:

1. $ y=\dfrac{x}{(1-x)^2(1+x)^3} $
2. $ y=\dfrac{x^p(1-x)^q}{1+x}$ , где $p,\ q$ — константы.

```python
from sympy import *
from IPython.display import display, Math

init_printing(use_latex='mathjax')

# Задача 1
x = symbols('x')
y1 = x / ((1 - x)**2 * (1 + x)**3)
y1_prime = cancel(diff(y1, x))

display(Math("Задача\ 1:"))
display(Math(f"y = {latex(y1)}"))
display(Math(f"y' = {latex(y1_prime)}"))

# Задача 2
p, q = symbols('p q', real=True)
y2 = x**p * (1 - x)**q / (1 + x)
y2_prime = factor(diff(y2, x))

display(Math("Задача\ 2:"))
display(Math(f"y = {latex(y2)}"))
display(Math(f"y' = {latex(y2_prime)}"))
```

## Задача 2
Найдите $f'(1)$, если

$ f(x)=x+ (x-1) \arcsin \left(\sqrt{\frac{x}{x+1}}\right)$

с помощью графика (постройте на графике две пересекающихся прямых в необходимой точке) и с помощью аналитического решения.

```python
from sympy import *
from IPython.display import display, Math
import numpy as np, matplotlib.pyplot as plt

# ── Аналитика ──
x = symbols('x', positive=True)
f = x + (x-1)*asin(sqrt(x/(x+1)))
f_prime = diff(f, x)
f1 = simplify(f_prime.subs(x, 1))  # 1 + π/4

display(Math("f'(x) = " + latex(f_prime)))
display(Math("f'(1) = " + latex(f1) + f" \\approx {float(N(f1)):.4f}"))

# ── График: две пересекающихся прямых в точке касания ──
xs = np.linspace(0.1, 2, 400)
ys = xs + (xs-1)*np.arcsin(np.sqrt(xs/(xs+1)))
plt.plot(xs, ys, 'b', label='$f(x)$')
plt.plot(xs, 1 + float(N(f1))*(xs-1), 'r--', label='касательная $y=f(1)+f\'(1)(x-1)$')
plt.axvline(1, color='gray', ls=':', alpha=0.7, label='вертикаль $x=1$')
plt.plot(1, 1, 'go', ms=8)
plt.title('Две прямые пересекаются в точке $(1,1)$'); plt.grid(alpha=0.3); plt.legend(); plt.show()
```

## Задача 3
Найдите производную для неявной функции и её значение при $x=2,\ y=4$.

$ x^2 + 2xy - y^2 - 2x = 0$

Найдите касательную к кривой в точке $x=2,\ y=4$ и сравните угол наклона касательной с полученной производной.

```python
from sympy import *
from IPython.display import display, Math
import numpy as np, matplotlib.pyplot as plt

# ── Аналитика ──
x, y = symbols('x y')
F = x**2 + 2*x*y - y**2 - 2*x
y_prime = simplify(-diff(F, x) / diff(F, y))

display(Math("y' = -\\frac{F_x}{F_y} = " + latex(y_prime)))
display(Math("y'(2,4) = \\frac{5}{2}"))

# ── График ──
def F_num(x, y): return x**2 + 2*x*y - y**2 - 2*x
xs = np.linspace(-1, 5, 400)
ys = np.linspace(-1, 7, 400)
X, Y = np.meshgrid(xs, ys)

plt.figure(figsize=(8,6))
plt.contour(X, Y, F_num(X, Y), levels=[0], colors='b', linewidths=2)
plt.plot(xs, 2.5*xs - 1, 'r--', lw=2, label='касательная $y=\\frac{5}{2}x-1$')
plt.plot(2, 4, 'go', ms=10, label='точка $(2,4)$')
plt.grid(alpha=0.3); plt.axhline(0, c='k', lw=0.5); plt.axvline(0, c='k', lw=0.5)
plt.legend(); plt.title('Неявная кривая и касательная'); plt.axis('equal')
plt.show()

display(Math("\\text{Угол наклона касательной совпадает с } y'(2,4)=\\frac{5}{2}"))
```

## Задача 4
Закон движения точки по оси $0x$ задаётся формулой

$x=10t+5t^2$, где $t$ — время в секундах, а $x$ — расстояние в метрах.

Найдите мгновенную скорость движения, когда $t=20$. Вычислите значение численно и аналитически, когда $\Delta t = 0.1$, $\Delta t = 0.01$. Сравните ошибку между аналитическим и численным решением.

```python
from sympy import *
from IPython.display import display, Math

# ── Аналитика ──
t = symbols('t', real=True)
x = 10*t + 5*t**2
v = diff(x, t)  # v = dx/dt
v20 = v.subs(t, 20)

display(Math("v(t) = \\frac{dx}{dt} = " + latex(v)))
display(Math("v(20) = " + latex(v20) + " \\text{ м/с}"))

# ── Численный расчёт ──
def x_num(t): return 10*t + 5*t**2

for dt in [0.1, 0.01]:
    v_num = (x_num(20 + dt) - x_num(20)) / dt
    error = abs(v_num - 210)
    display(Math(f"\\Delta t = {dt}: \\quad v_{{\\text{{числ}}}} = {v_num:.6f}, \\quad \\text{{ошибка}} = {error:.6f}"))
```

## Задача 1
Найдите неопределённый интеграл

$ \int x^2 dx$

и определённый интеграл

$ \int_{-1}^{2} x^2 dx$.

Найдите численно интеграл этой функции и сравните результат, $\Delta x = 0.01$.

```python
from sympy import *
from IPython.display import display, Math

# ── Аналитика ──
x = symbols('x')
F = integrate(x**2, x)          # неопределённый интеграл
I_exact = integrate(x**2, (x, -1, 2))  # определённый интеграл

display(Math("\\int x^2 \\, dx = " + latex(F) + " + C"))
display(Math("\\int_{-1}^{2} x^2 \\, dx = " + latex(I_exact)))

# ── Численный расчёт (метод трапеций) ──
import numpy as np
dx = 0.01
xs = np.arange(-1, 2 + dx, dx)
ys = xs**2
I_num = dx * (0.5*ys[0] + ys[1:-1].sum() + 0.5*ys[-1])
error = abs(I_num - float(I_exact))

display(Math(f"\\text{{Численно (}}\\Delta x={dx}\\text{{): }} \\int_{{-1}}^{{2}} x^2 dx \\approx {I_num:.6f}"))
display(Math(f"\\text{{Ошибка: }} |{I_num:.6f} - 3| = {error:.6f}"))
```

## Задача 2

Найдите частные производные для $f = e^x(\cos(y) + x \sin(y))$.

Также постройте график функции в 3D. Подберите адекватные ограничения по осям x и y. Постройте график в 2D, укажите градиент, а также его компоненты в точках $(0,0)$, $(5,1.8)$, $(5.5,-2)$.

```python
from sympy import *
from IPython.display import display, Math
import numpy as np, matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# ── Аналитика ──
x, y = symbols('x y')
f = exp(x) * (cos(y) + x*sin(y))

fx = diff(f, x)  # ∂f/∂x
fy = diff(f, y)  # ∂f/∂y

display(Math("f(x,y) = " + latex(f)))
display(Math("\\frac{\\partial f}{\\partial x} = " + latex(simplify(fx))))
display(Math("\\frac{\\partial f}{\\partial y} = " + latex(simplify(fy))))

# ── Градиент в точках ──
points = [(0, 0), (5, 1.8), (5.5, -2)]
for pt in points:
    gx = N(fx.subs({x: pt[0], y: pt[1]}))
    gy = N(fy.subs({x: pt[0], y: pt[1]}))
    display(Math(f"\\nabla f{pt} = ({gx:.2f},\\; {gy:.2f})"))

# ── 3D график (ограничения подобраны для читаемости) ──
fig = plt.figure(figsize=(12, 5))

# 3D поверхность
ax1 = fig.add_subplot(121, projection='3d')
xs = np.linspace(-1.5, 1.5, 100)  # ограничение по x: экспонента растёт быстро
ys = np.linspace(-2*np.pi, 2*np.pi, 100)
X, Y = np.meshgrid(xs, ys)
Z = np.exp(X) * (np.cos(Y) + X * np.sin(Y))
ax1.plot_surface(X, Y, Z, cmap='viridis', alpha=0.9)
ax1.set_title('3D: $f(x,y)=e^x(\\cos y + x\\sin y)$')
ax1.set_xlabel('x'); ax1.set_ylabel('y'); ax1.set_zlabel('f')

# 2D контур + градиенты
ax2 = fig.add_subplot(122)
contour = ax2.contourf(X, Y, Z, 20, cmap='viridis')
plt.colorbar(contour, ax=ax2)
ax2.set_title('Контур + градиенты')
ax2.set_xlabel('x'); ax2.set_ylabel('y')

# Векторы градиента в указанных точках
for pt in points:
    gx = float(fx.subs({x: pt[0], y: pt[1]}))
    gy = float(fy.subs({x: pt[0], y: pt[1]}))
    scale = 0.15 / max(np.hypot(gx, gy), 1e-5)  # масштабируем для визуализации
    ax2.quiver(pt[0], pt[1], gx*scale, gy*scale, 
               angles='xy', scale_units='xy', scale=1, 
               color='red', width=0.005, label=f'∇f{pt}' if pt==points[0] else "")
    ax2.plot(pt[0], pt[1], 'wo', markersize=6)

ax2.legend(loc='upper right')
ax2.axis('equal')
plt.tight_layout()
plt.show()
```

1. Найдите прямую кинематику для данного механизма (обратите внимание на финальную систему координат).
2. Посчитайте матрицу Якоби.

```python
from sympy import *
from IPython.display import display, Math

# 4 поступательных сочленения
q1, q2, q3, q4 = symbols('q1 q2 q3 q4')

# 1. Прямая кинематика в базовой СК
x_w = q1 + q3
y_w = q2 + q4

# 2. Преобразование в финальную СК (поворот на 90° против часовой стрелки)
# Правильная матрица поворота: [[0, 1], [-1, 0]]
X_e = y_w
Y_e = -x_w

# 3. Матрица Якоби
J_base = Matrix([
    [diff(x_w, q1), diff(x_w, q2), diff(x_w, q3), diff(x_w, q4)],
    [diff(y_w, q1), diff(y_w, q2), diff(y_w, q3), diff(y_w, q4)]
])

# Правильная матрица поворота
R = Matrix([[0, 1], [-1, 0]])
J_final = R * J_base

# Вывод
display(Math("\\text{Прямая кинематика:}"))
display(Math(f"X_e = {latex(X_e)}"))
display(Math(f"Y_e = {latex(Y_e)}"))

display(Math("\\text{Матрица Якоби (финальная СК):}"))
display(Math(latex(J_final)))
```

# Видео 6

## Задача 1
Дан маятник Капицы и его дифференциальное уравнение

${\ddot {\varphi }}=-(a\nu ^{2}\cos \nu t+g){\frac {\sin \varphi }{l}}$
, где:
- $\nu$ — частота вынуждающих вертикальных гармонических колебаний подвеса,
- $a$ — амплитуда вынуждающих колебаний,
- $\omega _{0}={\sqrt {g/l}}$ — собственная частота колебаний математического маятника,
- $g$ — ускорение свободного падения,
- $l$ — длина стержня,
-$m$ — масса грузика.

Зависимость координат грузика от времени:
$\left\{{\begin{aligned}x&=l\sin \varphi ,\\y&=-l\cos \varphi -a\cos \nu t.\end{aligned}}\right.$



1. Найдите уравнение движения маятника (численное решение).
2. Постройте графики $y(x)$, $y(t)$, $\dot \varphi(\varphi)$, когда
$a = 0.1$,
$\nu = 25$,
$l = 1$,
$m = 1$,
$\varphi_0 = \pi$,
$\dot \varphi_0 = 0.1$.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Параметры задачи
a = 0.1       # амплитуда вынуждающих колебаний подвеса, м
nu = 25       # частота вынуждающих колебаний, рад/с
l = 1.0       # длина стержня, м
g = 9.81      # ускорение свободного падения, м/с^2
phi0 = np.pi  # начальный угол, рад
dphi0 = 0.1   # начальная угловая скорость, рад/с

# Уравнение движения: система первого порядка
def pendulum_ode(t, u):
    phi, dphi = u
    d2phi = -(a * nu**2 * np.cos(nu * t) + g) / l * np.sin(phi)
    return [dphi, d2phi]

# Временной интервал
t_start, t_end = 0.0, 10.0  # секунды
t_eval = np.linspace(t_start, t_end, 5000)

# Начальные условия
u0 = [phi0, dphi0]

# Численное интегрирование
sol = solve_ivp(pendulum_ode, [t_start, t_end], u0, t_eval=t_eval, method='RK45', rtol=1e-9, atol=1e-12)

# Извлечение решения
t = sol.t
phi = sol.y[0]
dphi = sol.y[1]

# Координаты x, y
x = l * np.sin(phi)
y = -l * np.cos(phi) - a * np.cos(nu * t)

# Графики
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# 1. Траектория y(x)
axes[0, 0].plot(x, y, 'b', linewidth=0.5)
axes[0, 0].set_xlabel('x, м')
axes[0, 0].set_ylabel('y, м')
axes[0, 0].set_title('Траектория грузика y(x)')
axes[0, 0].grid(True)
axes[0, 0].set_aspect('equal')

# 2. Зависимость y(t)
axes[0, 1].plot(t, y, 'g', linewidth=0.5)
axes[0, 1].set_xlabel('t, с')
axes[0, 1].set_ylabel('y, м')
axes[0, 1].set_title('Вертикальная координата y(t)')
axes[0, 1].grid(True)

# 3. Фазовый портрет: dphi(phi)
axes[1, 0].plot(phi, dphi, 'r', linewidth=0.5)
axes[1, 0].set_xlabel('φ, рад')
axes[1, 0].set_ylabel('dφ/dt, рад/с')
axes[1, 0].set_title('Фазовый портрет dφ/dt(φ)')
axes[1, 0].grid(True)

# 4. Угол φ(t) для справки
axes[1, 1].plot(t, phi, 'm', linewidth=0.5)
axes[1, 1].set_xlabel('t, с')
axes[1, 1].set_ylabel('φ, рад')
axes[1, 1].set_title('Угол φ(t)')
axes[1, 1].grid(True)

plt.tight_layout()
plt.show()
```
